```python
import numpy as np
import damask
result_file = "../IMI817_34um_tensionY.hdf5"
```

```python
result = damask.Result(result_file)
```

```python
result
```

```python
result = result.view(increments=-1)
result
```

```python
stress = result.get("sigma_vM")
stress
```

```python
strain = result.get('epsilon_V^0.0(F)_vM')
strain
```

```python
texture = result.get("O")
print(texture.shape)
texture
```

```python
texture_quaternions = damask.Orientation.from_quaternion(q=texture, lattice="hP")
texture_quaternions
```

```python
euler = texture_quaternions.as_Euler_angles()
euler
```

```python
texture_euler = damask.Rotation(texture).as_Euler_angles()
texture_euler
```

```python
original_data_file = "../Preprocessing/IMI817_34um_coarse_clean.ang"
header_lines_in_original = 26
export_file_name = "../PostDeformedTexture.ang"
IQ_index = 5
CI_index = 6
```

```python
header = []
data_lines = []
with open(original_data_file, "r") as file:
    all_data = file.readlines()
    for i in range(header_lines_in_original):
        header.append(all_data[i])
    for i in range(header_lines_in_original, len(all_data)):
        data_lines.append(all_data[i].split())
print(header)
```

```python
data_lines
```

```python
# Updating data lines
for i, data in enumerate(data_lines):
    for j in range(3):
        data[j] = str(round(texture_euler[i][j], 6))
    data[IQ_index] = str(round(strain[i], 6))
    data[CI_index] = str(round(stress[i], 3))
data_lines
```

```python
# Save the data as ang
with open(export_file_name, "w") as write_file:
    for line in header:
        write_file.write(line)
    for data in data_lines:
        line = ""
        for datas in data:
            line += datas
            line += "\t"
        write_file.write(line + "\n")
```
